```python
import pandas as pd
import os
pwd = os.getcwd()
df = pd.read_excel(r"C:\Users\Elieen\Desktop\prob_5_data.xlsx")
target = "Buying"
print(df)
```

```python
X_data = df.drop(columns=[target,'User id'], inplace=False)
y_data = df[target]
X_pre=X_data.iloc[-1:,:]
X_train = X_data.iloc[:-1,:]
y_train = y_data.iloc[:-1]
print(X_train)
print(X_pre)
```

```python
X_features = X_data.columns.values
y_features =['0','1'] # 0--no, 1--yes
print(X_features)
```

```python
import numpy as np
import pandas as pd
import graphviz 
from sklearn import tree
from sklearn.datasets import load_wine
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
```

```python
def plot_tree(clf, feature_names, target_names):
    dot_data = tree.export_graphviz(clf, out_file=None, 
                     feature_names=feature_names,  
                     class_names=target_names,  
                     filled=True, rounded=True,  
                     special_characters=True)  
    graph = graphviz.Source(dot_data)  
    return graph
plot_tree(clf,X_features,y_features)
```

```python
clf.predict(X_pre)
```

```python
print("train_score", clf.score(X_train, y_train))
```
